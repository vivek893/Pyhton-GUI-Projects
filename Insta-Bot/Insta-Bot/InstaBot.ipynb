2.Type for “food” in search bar and print all the names of the Instagram Handles that are displayed in list after typing “food”<br>Note : Make sure to avoid printing hashtags

```python
def search(s):
    try:
        #locating serch bar and sending text)
        search_box = wait.until(EC.presence_of_element_located((By.XPATH,'//div[contains(@class,"xjoudau")]/input')))
        #search_box=driver.find_element('class name','x1lugfcp')
        
        search_box.send_keys(s)
        #waiting till all searched is located
        res = wait.until(EC.presence_of_element_located((By.XPATH,'//div[contains(@class,"x6s0dn4")]/div/a')))
        #extracting all serched handle
        handle_names=driver.find_element('class name','x6s0dn4')
        names=[]
        for i in handle_names:
            if i.text[0]!='#':
                names.append(i.text.split('\n')[0])
                print(names)
        time.sleep(5)
        #clearing search bar
        driver.find_element('class name','_aawn').click()
        return names
    except TimeoutException :
        print('No Search Found!')

```

Follow/Unfollow given handle - <br>

 2. Start following it. Print a message if you are already following<br>
 3. After following, unfollow the instagram handle. Print a message if you have already unfollowed.<br>

```python
def follow():
    try:
        find=wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_acan')))
        if find.text=='Follow':
            find.click()
            time.sleep(3)
        else:
            print('Already Follow')
           
    except TimeoutException :
        print('something went wrong')
```

```python
def unfollow():
    try:
        find=wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_acan')))
        if find.text!='Follow':
            find.click()
            time.sleep(3)
            #locating popup window (when you click on follow button)
            c=driver.find_element(By.XPATH,'//div/div/div/div/div/span[@dir="auto"]')
            c.click()
            time.sleep(2)
        else:
            print('Already Unfollowed')
        # add to in close friend list
        b=driver.find_element('class name','x7r02ix')
        print('you want to add that profile in close friend list: if Yes then press:1 otherwise press any things')
        x=int(input())
        if x==1:
            b=driver.find_element('class name','x7r02ix')
            b.click()
        # if remove from close friend list:
        print('if you want to remove from close friend list then press:0 otherwise press anything')
        y=int(input())
        if y==0:
            b.click()
        
        
    except TimeoutException :
        print('something went wrong')
```

```python
def Check_Story():
    try:
        #locating story or profile pic
        story = wait.until(EC.presence_of_element_located((By.CLASS_NAME,"x4afe7t")))
    
        #check the Profile photo size to find out story is available or not
        height = driver.find_element('class name','xpdipgo').get_attribute('height')
        if int(height) == 166:
            print("Already seen the story")
        else:
            print("Viewing the story")
            driver.execute_script("arguments[0].click();",story)
    except:
        print("No story is available to view")
```
